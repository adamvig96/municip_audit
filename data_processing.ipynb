### Randomizing gender

```python
random.seed(42)
genders = ["male", "female"]
data["gender"] = [random.choice(genders) for i in range(data.shape[0])]
```

### String formating for solving a(z) and -i problem

```python
vowels = ["A","Á","E","É","I","Í","O","Ó","Ö","Ő","U","Ú","Ü","Ű"]

consonants = ["B","C","D","F","G","H","J","K","L","M","N","P","R","S","T","V","W","Z"]
```

```python
data["formated_town"] = None

for i in range(data.shape[0]):
    if data.loc[i, "request"] in ["Cemetary", "Nursery"]:
        if data.loc[i, "town"][-1] == "i":
            data.loc[i, "formated_town"] = data.loc[i, "town"]
        else:
            data.loc[i, "formated_town"] = data.loc[i, "town"] + "i"

        if data.loc[i, "town"][0] in vowels:
            data.loc[i, "formated_town"] = "z " + data.loc[i, "formated_town"]
        elif data.loc[i, "town"][0] in consonants:
            data.loc[i, "formated_town"] = " " + data.loc[i, "formated_town"]
        else:
            print("jaj")
            break

    elif data.loc[i, "request"] in ["Camping", "Wedding"]:
        data.loc[i, "formated_town"] = data.loc[i, "town"]
    else:
        print("jaj")
        break
```

### dividing waves to 7 sending days
counties represented in each day as in the whole wave

```python
counties = list(set(data["county.x"]))

data["day"] = None
for item in counties:
    county = data.loc[data["county.x"] == item]
    splitted = np.array_split(county, 7)
    for day, split in enumerate(splitted):
        for j in split.index:
            data.loc[j, "day"] = day + 1
```
